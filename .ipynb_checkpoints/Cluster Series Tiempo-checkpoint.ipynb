# Clustering en series de tiempo

Una vez seleccionada la medida de distancia (depende del propósito de agrupamiento deseado), se puede obtener una matriz D de disimilaridades por pares y utilizar esta matriz como punto de partida de un algoritmo de segmentación clásico. La segmentación jerárquica es un algoritmo que intenta crear una jerarquía de grupos en la que, a medida que aumenta el nivel de la jerarquía, los grupos se crean al fusionar los grupos del siguiente nivel inferior; para decidir cómo se realiza la fusión, se debe especificar una medida de similaridades entre grupos, además de la que se utiliza para calcular las similaridades por pares. Existen dos tipos de metodologías para la segmentación jerárquica: el aglomerativo y el divisivo. En la metodología aglomerativa, cada dato comienza en su propio grupo, y los miembros se agrupan secuencialmente en función de la medida de disimilaridad hasta que todos los miembros estén contenidos en un solo grupo. Los procedimientos divisivos hacen exactamente lo contrario, comenzando con todos los datos en un grupo y dividiéndolos hasta que cada miembro esté en un grupo.

En general, los CVI se pueden adaptar a particiones duras o suaves. Para los primeros, los CVI se pueden clasificar como internos, externos o relativos según cómo se calculen. Los CVI internos solo consideran los datos particionados e intentan definir una medida de pureza del grupo, mientras que los CVI externos comparan la partición obtenida con la correcta.  

El índice de silueta es un ejemplo de un CVI interno, definiendo el valor de silueta para $x_i$ dado el agrupamiento $C_K$ como
$$s_i(C_K) = s_{iK} =\frac{b_i-a_i}{max(a_i,b_i)}$$
donde $a_i$ es la disimilaridad promedio del objeto xi a los
restantes objetos dentro de el mismo cluster, $b_i=min_{c\neq C(i)} d_i^C$ con $d_i^C$ es la disimilaridad promedio del objeto $x_i$ a los objetos que están en clusters diferentes de su mismo cluster.
Por otra parte, un CVI externo se basa en calcular la variaciación entre la verdadera segmentación G y la solución del método aplicado A. El indice de dimilaridad $Sim\,(G,A)$ se define como:
$$Sim\,(G,A)=\frac{1}{k}\sum_{i=1}^{k}max_{i\leq j\leq k}Sim(G_i,A_j)$$
$$Sim(G_i,A_j)=\frac{|G_i\cap A_j|}{|G_i|+|A_j|}$$
