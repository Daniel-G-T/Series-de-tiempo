# **Introducción a series de tiempo**

Una serie de tiempo se puede definir como una colección de variables aleatorias indexadas por el tiempo y se denota como $\{X_{t}\}_{t \in \mathbb{Z}}$ donde $X_t$ es una variable aleatoria en el tiempo $t$.

Al graficar una serie de tiempo, los valores de las variables aleatorias se establecen en el eje vertical, u ordenadas, con la escala de tiempo como la abscisa. Además, es común conectar los valores en períodos de tiempo adyacentes para reconstruir visualmente una serie de tiempo continua hipotética que podría haber producido esos valores.

## Ejemplos de series de tiempo

### **Ruido blanco Gausiano**

Una serie de tiempo construida con un conjunto de variables aleatorias no correlacionadas,$w_t$, con media 0 y varianza finita $\sigma_w^2$, se llama ruida blanco; este proceso se denota como $w_t\sim wn(0,\sigma_w^2)$.

Una serie de tiempo de ruido blanco con variables independientes normales con media 0 y varianza $\sigma_w^2$, se llama ruida blanco Gaussiano; este proceso se denota como $w_t\sim iidN(0,\sigma_w^2)$. En la Figura \ref{fig:wn} se muestra la serie de $w_t\sim iidN(0,1)$.

```python
state = check_random_state(0)
w_t=state.normal(loc=0,scale=1,size=200)
```

### **Caminata aleatoria con deriva**

Supongamos que tenemos una serie de ruido blanco $w_t$ con condición inicial $x_0=0$ y una deriva constante $\delta$, el modelo esta dado por
$$x_t=\delta+x_{t-1}+w_t\;\;\;\;\;t=1,2,\dots$$

```python
delta=0.5
x_t=np.concatenate([[0],np.cumsum(delta+w_t)])
```

### **Señal con ruido**
Sea $w_t$ una serie de ruido blanco, muchos modelos realistas para generar series de tiempo asumen una señal contaminada con un ruido; esto se puede escribir como
$$x_t=Acos(2\pi\omega t + \phi)+w_t\;\;\;\;\;t=1,2,\dots,$$
donde A es la amplitud, $\omega$ es la frecuencia de oscilación y $\phi$ es el desplazamiento de fase. A continuación se muestra la señal $2cos(\frac{2\pi t}{50} + 0.6\pi)$ y con un ruido blanco.

```python
t=np.linspace(start=1, stop=w_t.shape[0],num=w_t.shape[0])
y_t=2*np.cos((2*np.pi*t/50)+(0.6*np.pi))+w_t
```

## Medidas de distancias

Las medidas de distancia proporcionan una cuantificación de la disimilitud entre dos datos. Uno de los aspectos más importantes  de los algoritmos de segmentación es calcular distancias entre los distintos objetos del conjunto. Muchos tipos de medidas de disimilaridad han sido propuestas para series de tiempo, las cuales se han agrupado en cuatro categorías: medidas libres de modelo, medidas basadas en modelo, medidas basadas en complejidad y medidas basadas en predicción.

### Medidas libre de modelo

Un enfoque para medir la proximidad entre dos series de tiempo, $X_t$ e $Y_t$, es considerar métricas convencionales basadas en la cercanía de sus valores en puntos específicos de tiempo. Algunas medidas de este tipo son:
- **Distancia de Minkowski** La distancia de Minkowski de orden $q \in \mathbb{Z}^+$, norma $L_{q}$, $$d_{L_q}(\{X_t\},\{Y_t\})=(\sum_{t=1}^T(X_t-Y_t)^q)^{\frac{1}{q}}$$ La distancia Euclideana es un caso particular de esta distancia de Minkowski(q=2). En la Figura \ref{fig:d_ecu} se muestra el cómo se crea la asociación entre los valores de las dos series para calcular la diferencia y calcular la distancia.

- **Distancia de Fréchet:** Sea M el conjunto de todos los posibles pares preservando el orden de las observaciones con la forma $$r=((X_{a_1},Y_{b_1}),\dots,(X_{a_m},Y_{b_m}))$$ con $a_i,b_j\in \{1,\dots,T\}$ tal que $a_1=b_1=1$, $a_m=b_m=T$, $a_{i+1}=a_i$ o $a_{i+1}=a_i+1$ y $b_{i+1}=b_i$ o $b_{i+1}=b_i+1$ para $i\in\{1,\dots,m-1\}$. Entonces la distancia de \textit{Fréche} se define como: $$d_F(\{X_t\},\{Y_t\})=min_{r\in M}\Big(max_{i=1,\dots,m}|X_{a_i}-Y_{b_i}|\Big)$$
    
- **Distancia de la deformación dinamica de tiempo:** La metodología de la deformación dinámica de tiempo (DTW) se basa en deformar el eje del tiempo con el objetivo de minimizar la diferencia acumulada entre los puntos alineados al considerar todas las posibles compresiones o desplazamientos respecto a la variable del tiempo. Sean $\{X_t\}$,$\{Y_t\}$ dos series de tiempo con longitud N y M, respectivamente, entonces la curva de deformación $W=(\phi_t, \psi_t)$ con $t=1,\dots,T$ nos permite relacionar una serie con la otra mapeando los índices a través de $\phi_t$ y $\psi_t$.
La curva de deformación está sujeta a las siguientes restricciones:
1. Restricción de punto inicial: $$\phi_1 = \psi_1 = 1$$
2. Restricción de punto final: $$\phi_T = N,\;\; \psi_T = M$$
3. Monotonicidad: $$\phi_t \geq \phi_{t-1}$$ $$\psi_t \geq \psi_{t-1}$$

La curva de deformación óptima $\hat{W}$ esta dada por
$$\hat{W}= (\hat{\phi}_t,\hat{\psi}_t)=arg\;min_{\phi_t,\psi_t}\sum_{t=1}^{T} d(x_{\phi_t},y_{\psi_t})$$

**Ejemplo**

```python
x = [3, 1, 2, 2, 1]
y = [2, 0, 0, 3, 3, 1, 0]
W=[(0, 0), (1, 1), (1, 2), (2, 3), (3, 4), (4, 5), (4, 6)]
```
